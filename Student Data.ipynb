```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Read the CSV file into a DataFrame
df = pd.read_csv('Fake Student Data.csv')
df.head()
```

```python
df.info()
```

```python
# Step 2: Calculate average GPA by gender
avg_gpa_by_gender = df.groupby('Gender')['GPA'].mean().reset_index()

# Step 3: Create a bar plot
plt.figure(figsize=(8, 6))  # Adjust the figure size (width, height)

# Plotting the bars
plt.bar(avg_gpa_by_gender['Gender'], avg_gpa_by_gender['GPA'], color=['pink', 'blue'])

# Adding titles and labels
plt.title('Average GPA by Gender')
plt.xlabel('Gender')
plt.ylabel('Average GPA')

# Display the plot
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```

```python
# Step 4: Calculate average Absent Da7s by gender
avg_gpa_by_gender = df.groupby('Gender')['Number of Days Absent'].mean().reset_index()

# Step 5: Create a bar plot
plt.figure(figsize=(8, 6))  # Adjust the figure size (width, height)

# Plotting the bars
plt.bar(avg_gpa_by_gender['Gender'], avg_gpa_by_gender['Number of Days Absent'], color=['pink', 'blue'])

# Adding titles and labels
plt.title('Average Number of Days Absent')
plt.xlabel('Gender')
plt.ylabel('Number of Days Absent')

# Display the plot
plt.tight_layout()  # Adjust layout for better fit
plt.show()
```

```python
# Step 6: Calculate average GPA and days absent by gender
avg_data = df.groupby('Gender').agg({'GPA': 'mean', 'Number of Days Absent': 'mean'}).reset_index()

# Step 7: Create a grouped bar plot
plt.figure(figsize=(10, 6))  # Adjust the figure size (width, height)

# Bar width
bar_width = 0.35

# Positions of bars on X axis
r1 = range(len(avg_data)) 

# Plotting bars
plt.bar(r1, avg_data['GPA'], width=bar_width, edgecolor='grey', label='GPA')
plt.bar([x + bar_width for x in r1], avg_data['Number of Days Absent'], width=bar_width, edgecolor='grey', label='Number of Days Absent')

# Adding labels, title, and ticks
plt.xlabel('Gender', fontweight='bold')
plt.xticks([r + bar_width/2 for r in r1], avg_data['Gender'])
plt.ylabel('Average', fontweight='bold')
plt.title('Average GPA and Days Absent by Gender')
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Step 8: Visualize Absent Days to GPA by Gender
sns.lmplot(x='Number of Days Absent', y='GPA', hue='Gender', 
           data=df.loc[df['Gender'].isin(['M'])], 
           fit_reg=False)
plt.show()
```

```python
# Step 9: Visualize Absent Days to GPA by Gender
sns.lmplot(x='Number of Days Absent', y='GPA', hue='Gender', 
           data=df.loc[df['Gender'].isin(['F'])], 
           fit_reg=False)
plt.show()
```

```python
# Step 10: Visualize GPA by Gender and differentiate Grade level
sea = sns.FacetGrid(df, col = "Grade",
                    height = 4, aspect  =.5, hue='Gender', palette=['blue', 'red'])
 
sea.map(sns.barplot, "Gender", "GPA",
        order = ["F", "M"])
```

```python
# Step 10: Visualize Days Absent by Gender and differentiate Grade level
sea = sns.FacetGrid(df, col = "Grade",
                    height = 4, aspect  =.5, hue='Gender', palette=['blue', 'red'])
 
sea.map(sns.barplot, "Gender", "Number of Days Absent",
        order = ["F", "M"])
```
